```python
precos = [1.55, 1.42, 1.33, 1.80]

for i in range(1, len(precos)):

    preco_atual = precos[i]
    preco_anterior = precos[i - 1]

    if preco_atual > preco_anterior:
        print("Vela", i, "SUBIU")

    else:
        print("Vela", i, "CAIU")
```

```python
precos = [1.55, 1.42, 1.33, 1.80]

for i in range(1, len(precos)):

    atual = precos[i]
    anterior = precos[i - 1]

    diferenca = atual - anterior

    if diferenca > 0.20:
        print("Vela", i, "SUBIU FORTE")

    elif diferenca < -0.20:
        print("Vela", i, "CAIU FORTE")

    else:
        print("Vela", i, "MOVIMENTO FRACO")
```

```python
def analisar_movimento(precos):

  for i in range(1, len(precos)):

    atual = precos[i]
    anterior = precos[i - 1]

    diferenca = atual - anterior

    if diferenca > 0.20:
      print("Vela", i, "Subiu Forte | Diferença:", diferenca)

    elif diferenca < -0.20:
      print("Vela", i, "Caiu Forte | Diferença:", diferenca)

    else:
      print("Vela", i, "Movimento Fraco | Diferença:", diferenca)


precos = [1.55, 1.42, 1.33, 1.80]

analisar_movimento(precos)

```
